* Resumo: Este trabalho tem como objetivo desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial, em que as salas de cinema poderão estar a exibir filmes ou não; cada sala tem uma determinada lotação; uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro; e cada sala tem um filme associado. Foi seguido o modelo sugerido de maneira a desempenhar as funções propostas.

```python
# Função para listar os filmes em exibição nas salas do cinema
def listar(cinema):
    if not cinema:
        print("Não há filmes em exibição.")
    else:
        for sala in cinema:
            print(f"Sala: {sala[2]} ({len(sala[1])} lugares ocupados)")

# Função para verificar se um lugar está disponível em uma sala
def disponivel(cinema, filme, lugar):
    for sala in cinema:
        if sala[2] == filme:
            return lugar not in sala[1]
    return False

# Função para vender um bilhete e ocupar um lugar na sala
def vendebilhete(cinema, filme, lugar):
    for sala in cinema:
        if sala[2] == filme and lugar not in sala[1]:
            sala[1].append(lugar)
    return cinema

# Função para listar as disponibilidades em cada sala do cinema
def listardisponibilidades(cinema):
    for sala in cinema:
        lugares_disponiveis = sala[0] - len(sala[1])
        print(f"Sala: {sala[2]} - Filme: {sala[2]}, Lugares disponíveis: {lugares_disponiveis}")

# Função para inserir uma nova sala no cinema
def inserirSala(cinema, sala):
    nomes_salas = [s[2] for s in cinema]
    if sala[2] not in nomes_salas:
        cinema.append(sala)
    return cinema

# Função para listar todos os filmes em exibição no cinema
def listarFilmesEmExibicao(cinema):
    filmes = set([s[2] for s in cinema])
    print("Filmes em exibição:")
    for filme in filmes:
        print(filme)

# Função para mostrar um menu de interface
def menu():
    cinema = []
    while True:
        print("\nMenu:")
        print("1. Listar filmes em exibição")
        print("2. Listar disponibilidades em cada sala")
        print("3. Inserir sala")
        print("4. Vender bilhete")
        print("5. Listar salas no cinema")
        print("0. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            listarFilmesEmExibicao(cinema)
        elif opcao == "2":
            listardisponibilidades(cinema)
        elif opcao == "3":
            sala = (int(input("Lotação da sala: ")), [], input("Nome do filme: "))
            cinema = inserirSala(cinema, sala)
        elif opcao == "4":
            nome_filme = input("Nome do filme: ")
            lugar = int(input("Número do lugar: "))
            if disponivel(cinema, nome_filme, lugar):
                cinema = vendebilhete(cinema, nome_filme, lugar)
                print("Bilhete vendido com sucesso.")
            else:
                print("O lugar já está ocupado.")
        elif opcao == "5":
            listar(cinema)
        elif opcao == "0":
            print("Obrigado por usar o sistema.")
            return
        else:
            print("Opção inválida. Tente novamente.")

menu()
```
