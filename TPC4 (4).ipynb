* Resumo: Desenvolver uma aplicação para manipulação de listas de números inteiros, que desempenha funções como soma, média, maior, menor, ordenar por ordem crescente e decrescente, procurar um elemento e sair.

```python
# Função para calcular a soma dos elementos da lista
def soma(lista):
    return sum(lista)

# Função para calcular a média dos elementos da lista
def media(lista):
    return sum(lista) / len(lista)

# Função para encontrar o maior elemento da lista
def maior(lista):
    return max(lista)

# Função para encontrar o menor elemento da lista
def menor(lista):
    return min(lista)

# Função para ordenar a lista em ordem crescente
def ordenar_crescente(lista):
    return sorted(lista)

# Função para ordenar a lista em ordem decrescente
def ordenar_decrescente(lista):
    return sorted(lista, reverse=True)

# Função para procurar um elemento na lista
def procurar_elemento(lista, elemento):
    if elemento in lista:
        return f'O elemento {elemento} foi encontrado na lista.'
    else:
        return f'O elemento {elemento} não foi encontrado na lista.'

# Função principal
def app():
    lista = []
    
    while True:
        print("\nOperações disponíveis:")
        print("1 - Adicionar número à lista")
        print("2 - Calcular a soma")
        print("3 - Calcular a média")
        print("4 - Encontrar o maior número")
        print("5 - Encontrar o menor número")
        print("6 - Ordenar a lista em ordem crescente")
        print("7 - Ordenar a lista em ordem decrescente")
        print("8 - Procurar um elemento na lista")
        print("9 - Sair")
        
        opcao = input("Escolha uma opção: ")
        
        if opcao == '1':
            numero = int(input("Digite um número inteiro: "))
            lista.append(numero)
        elif opcao == '2':
            print(f"A soma dos elementos da lista é: {soma(lista)}")
        elif opcao == '3':
            print(f"A média dos elementos da lista é: {media(lista)}")
        elif opcao == '4':
            print(f"O maior elemento da lista é: {maior(lista)}")
        elif opcao == '5':
            print(f"O menor elemento da lista é: {menor(lista)}")
        elif opcao == '6':
            print(f"Lista em ordem crescente: {ordenar_crescente(lista)}")
        elif opcao == '7':
            print(f"Lista em ordem decrescente: {ordenar_decrescente(lista)}")
        elif opcao == '8':
            elemento = int(input("Digite o elemento que deseja procurar: "))
            print(procurar_elemento(lista, elemento))
        elif opcao == '9':
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__app__":
    app()
```

```python
app()
```
